```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data
claire = pd.read_csv('/content/accidentsdata.csv')

# Perform one-hot encoding for the 'Involving_Drink_Driving' column
claire = pd.get_dummies(claire, columns=['Involving_Drink_Driving'], drop_first=True)

# Independent variable (X) and dependent variable (y)
X = claire[['Involving_Drink_Driving_Yes']]  # Yes=Involving Drink Driving
y = claire['Count_Crashes']

# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
# Plotting the data points
claire.plot(x='Involving_Drink_Driving_Yes', y='Count_Crashes', style='1', label='accidents cause')

# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=1, label='Regression Line')
plt.xlabel('Involving_Drink_Driving')
plt.ylabel('Count_Crashes')
plt.legend()
plt.show()
```

```python

# Use the model to predict Count_Crashes
predicted_severity = regressor.predict([[1]])
# Printing the predicted accident severity
print(f"Predicted Accident Severity (Count_Crashes) for Involving Drink Driving: {predicted_severity[0]:.2f}")
```
